```python
# File processing
import os

# Data processing
from tqdm import tqdm
import numpy as np
import pandas as pd
import random

# Plotting
import matplotlib.pyplot as plt

# Hic, Distance and Structure creation
from numpy import linalg as LAnumpy
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial import distance_matrix
```

# Constants

```python
NB_BINS = 202
NB_STRUCTURES_TRAINING = 800
NB_STRUCTURES_TESTING = 200
EXPONENT = 1
DIGITS_FORMAT = '{0:0=4d}'
SEED = 1
EMBEDDING_SIZE = 3

# Trussart analysis dataset constants
TRUSSART_HIC_PATH = '../../experiments/data/trussart/hic_matrices/150_TADlike_alpha_150_set0.mat'
TRUSSART_STRUCTURES_PATH = '../../experiments/data/trussart/structure_matrices/'
TRUSSART_ENSEMBLE_SIZE = 100
```

```python
NB_NORMALS_X = 4
NB_NORMALS_Y = 6
NB_NORMALS_Z = 1
```

# Seeds

```python
random.seed(SEED)
np.random.seed(SEED)
```

# Tool functions

```python
def compute_hic_matrix(distance_matrix, alpha):
    
    distance_matrix = np.where(distance_matrix == 0, np.inf, distance_matrix)
    
    hic_matrix = np.zeros((len(distance_matrix), len(distance_matrix)))
    hic_matrix = np.where(distance_matrix == np.inf, hic_matrix, np.power(distance_matrix, -alpha))
    
    return hic_matrix
```

```python
def normalize_numpy(z):
    
    norm = LAnumpy.norm(z, 2, axis=1)
    max_norm = np.max(norm, axis=0)
    if max_norm == 0:
        max_norm = 1
    
    return z / max_norm
```

```python
def centralize_numpy(z):
    return z - np.mean(z, axis=0)
```

```python
def centralize_and_normalize_numpy(z):
    
    # Translate
    z = centralize_numpy(z)
    
    # Scale
    z = normalize_numpy(z)
    
    return z
```

```python
def normal(x, mu, sigma, amplitude):
    return np.asarray(amplitude * np.exp( - (x - mu)**2 / (2 * sigma**2)))
```

```python
def get_amplitude_signal(nb_normals):
    
    x = np.arange(NB_BINS)
    amplitude_signal = np.zeros((NB_BINS))
    mus = []
    sigmas = []
    amplitudes = []
    for i in range(nb_normals):
        mu = np.random.uniform(0, NB_BINS)
        sigma = np.random.uniform(2, 5)
        amplitude = np.random.uniform(0.002, 0.004)
        
        amplitude_signal += normal(x, mu, sigma, amplitude) 
        
        mus.append(mu)
        sigmas.append(sigma)
        amplitudes.append(amplitude)
        
    mus = np.asarray(mus)
    sigmas = np.asarray(sigmas)
    amplitudes = np.asarray(amplitudes)
        
    return amplitude_signal, mus, sigmas, amplitudes
```

```python
def get_amplitude_structure():
    
    amplitude_structure = np.zeros((NB_BINS, EMBEDDING_SIZE))
    amplitude_structure[:, 0], mus_x, sigmas_x, amplitudes_x = get_amplitude_signal(NB_NORMALS_X)
    amplitude_structure[:, 1], mus_y, sigmas_y, amplitudes_y = get_amplitude_signal(NB_NORMALS_Y)
    amplitude_structure[:, 2], mus_z, sigmas_z, amplitudes_z = get_amplitude_signal(NB_NORMALS_Z)
    
    mus = np.hstack((np.hstack((mus_x, mus_y)), mus_z))
    sigmas = np.hstack((np.hstack((sigmas_x, sigmas_y)), sigmas_z))
    amplitudes = np.hstack((np.hstack((amplitudes_x, amplitudes_y)), amplitudes_z))
    
    return amplitude_structure, mus, sigmas, amplitudes
```

```python
def get_random_sample(original_structure, amplitude_structure):

    # Parameter of the uniform distribution
    theta = np.sqrt(12 * amplitude_structure)
    u = np.random.uniform(-1, 1) * 0.5 * theta
    
    sample = original_structure + u
    return sample
```

```python
def get_random_samples(original_structure, nb_samples):
    
    amplitude_structure, mus, sigmas, amplitudes = get_amplitude_structure()
    
    samples = np.zeros((nb_samples, NB_BINS, EMBEDDING_SIZE))
    for i in range(nb_samples):
        samples[i,:,:] = get_random_sample(original_structure, amplitude_structure)
        
    return samples, mus, sigmas, amplitudes
```

## Structure matrices

```python
trussart_structures = []

file_list = os.listdir(TRUSSART_STRUCTURES_PATH)
file_list = filter(lambda f: f.endswith('.xyz'), file_list)

for file_name in file_list:
    current_trussart_structure = np.loadtxt(TRUSSART_STRUCTURES_PATH + file_name, dtype='f', delimiter='\t')
    current_trussart_structure = current_trussart_structure[:,1:]
    current_trussart_structure = centralize_and_normalize_numpy(current_trussart_structure)
    trussart_structures.append(current_trussart_structure)
    
trussart_structures = np.asarray(trussart_structures)
```

```python
mu_x = mus[:NB_NORMALS_X]
mu_y = mus[NB_NORMALS_X:NB_NORMALS_X + NB_NORMALS_Y]
mu_z = mus[NB_NORMALS_X + NB_NORMALS_Y:]
    
sigma_x = sigmas[:NB_NORMALS_X]
sigma_y = sigmas[NB_NORMALS_X:NB_NORMALS_X + NB_NORMALS_Y]
sigma_z = sigmas[NB_NORMALS_X + NB_NORMALS_Y:]
    
amplitude_x = amplitudes[:NB_NORMALS_X]
amplitude_y = amplitudes[NB_NORMALS_X:NB_NORMALS_X + NB_NORMALS_Y]
amplitude_z = amplitudes[NB_NORMALS_X + NB_NORMALS_Y:]
    
predicted_signal_x = np.zeros(NB_BINS)
for i in range(NB_NORMALS_X):
    predicted_signal_x += normal(np.arange(NB_BINS), mu_x[i], sigma_x[i], amplitude_x[i])
    
predicted_signal_y = np.zeros(NB_BINS)
for i in range(NB_NORMALS_Y):
    predicted_signal_y += normal(np.arange(NB_BINS), mu_y[i], sigma_y[i], amplitude_y[i])

predicted_signal_z = np.zeros(NB_BINS)
for i in range(NB_NORMALS_Z):
    predicted_signal_z += normal(np.arange(NB_BINS), mu_z[i], sigma_z[i], amplitude_z[i])

predicted_signal = np.zeros((NB_BINS, EMBEDDING_SIZE))
predicted_signal[:,0] = predicted_signal_x
predicted_signal[:,1] = predicted_signal_y
predicted_signal[:,2] = predicted_signal_z
```

# Ensemble Data Generator

```python
# test = np.zeros((100, 202, 3))
```

```python
def ensemble_data_generator(is_training=True):
    
    if is_training:
        nb_structures = NB_STRUCTURES_TRAINING
        file_name = 'train'
    else:
        nb_structures = NB_STRUCTURES_TESTING
        file_name = 'test'
        
    # Create Dataset
    for i in tqdm(range(nb_structures)):

        # Set the seeds
        random.seed(i*3 + 500)
        np.random.seed(i*1201 + 8302)

        samples, mus, sigmas, amplitudes = \
                get_random_samples(trussart_structures[np.random.choice(range(TRUSSART_ENSEMBLE_SIZE))], 
                    TRUSSART_ENSEMBLE_SIZE)
        
        if i == 19:
            return samples
        
        # Mu matrix to file
        df = pd.DataFrame(data=mus.astype(float))
        df.to_csv('../data/ensemble/' + file_name + '/mu_matrices/ensemble_mu_'\
                  + DIGITS_FORMAT.format(i) + '.txt', sep=' ', header=False, index=False)
        
        # Sigma matrix to file
        df = pd.DataFrame(data=sigmas.astype(float))
        df.to_csv('../data/ensemble/' + file_name + '/sigma_matrices/ensemble_sigma_'\
                  + DIGITS_FORMAT.format(i) + '.txt', sep=' ', header=False, index=False)
        
        # Amplitude matrix to file
        df = pd.DataFrame(data=amplitudes.astype(float))
        df.to_csv('../data/ensemble/' + file_name + '/amplitude_matrices/ensemble_amplitude_'\
                  + DIGITS_FORMAT.format(i) + '.txt', sep=' ', header=False, index=False)
        
        # Compute the hic matrix 
        hic_matrices = np.zeros((TRUSSART_ENSEMBLE_SIZE, NB_BINS, NB_BINS))
        for j in range(TRUSSART_ENSEMBLE_SIZE):
            sample = samples[j]
            centralized_normalized_sample = centralize_and_normalize_numpy(sample)
            distance = distance_matrix(centralized_normalized_sample, centralized_normalized_sample)
            hic_matrix = compute_hic_matrix(distance, EXPONENT) 
            hic_matrices[j,:,:]  = hic_matrix
            
        hic_matrix = np.sum(hic_matrices, axis=0)
        scaler = MinMaxScaler()
        hic_matrix = scaler.fit_transform(hic_matrix)

        # HiC matrix to file
        df = pd.DataFrame(data=hic_matrix.astype(float))
        df.to_csv('../data/ensemble/' + file_name + '/hic_matrices/ensemble_hic_' + \
                  DIGITS_FORMAT.format(i) + '.txt', sep=' ', header=False, index=False)
```

### Training

```python
test = ensemble_data_generator()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(20, 10))
fig.tight_layout(pad=2.0)
axis_names = ['X','Y','Z']

# Plot prediction
for j in range(EMBEDDING_SIZE):


    axs[j].set_xlim(0, NB_BINS)
    axs[j].set_ylim(-1, 1)
    axs[j].tick_params(axis='both', which='major', labelsize=20, width=4)

    if j == 2:
        axs[j].set_xlabel('Bins', fontsize=30)
    if j == 1:
        axs[j].set_ylabel('Variance', fontsize=30)

    for k in range(TRUSSART_ENSEMBLE_SIZE):
        axs[j].plot(range(NB_BINS), test[k,:,j], color='gray', 
              marker='o', linestyle='solid', linewidth=0.2, markersize=1, alpha=0.4)

#             plt.savefig('sample_per_axis_structure_variance.png')

plt.show()
```

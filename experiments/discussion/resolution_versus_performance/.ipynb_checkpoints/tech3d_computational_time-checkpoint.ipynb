# Plot

```python
x = [1, 20, 40, 80, 160, 320, 640, 1280]
y_data = [3, 5, 10, 22.75, 68.36, 253.70, 974.79, 4069.18]
y_network = [50, 160, 174.34, 242.34, 493.19, 1795.71, 2128.92, 5163]
```

```python
fig, axs = plt.subplots(1, 1, figsize=(20,11))

axs.scatter(x, y_data, color='red', s=200)
axs.plot(x, y_data, color='red', linestyle='dashed', linewidth=3, label=r'$log(Measured)$')
# axs.plot(np.asarray(range(1, 1280)), 
#          np.log(np.asarray(range(1, 1280))), 
#          color='green', linestyle='dashed', linewidth=3, label=r'$log(x)$')


axs.set_xlabel('Number of Bins', fontsize = 35)
axs.set_ylabel('Time (in s)', fontsize = 35)
axs.set_ylim(0, 5500)

axs.set_title('Time to generate synthetic data \n versus the number of bins of the HiC matrix', 
              size=40)

axs.tick_params(axis='both', which='major', labelsize=20, width=4)
plt.xticks(ticks=x, labels=x, rotation=75)
# plt.legend(fontsize=35, loc=(0.6, 0.25))

plt.savefig('time_generate_data_versus_number_bins.png')
plt.show()
```

```python
fig, axs = plt.subplots(1, 1, figsize=(20,11))

axs.scatter(x, y_network, color='red', s=200)
axs.plot(x, y_network, color='red', linestyle='dashed', linewidth=3, label=r'$log(Measured)$')

# axs.plot(np.asarray(range(1, 1280)), 
#          np.log(np.asarray(range(1, 1280))), 
#          color='green', linestyle='dashed', linewidth=3, label=r'$log(x)$')


axs.set_xlabel('Number of Bins', fontsize = 45)
axs.set_ylabel('Time (in s)', fontsize = 45)
axs.set_ylim(0, 5500)

axs.set_title('Time to train the linear network on synthetic data \n versus the number of bins of the HiC matrix', 
              size=45)

axs.tick_params(axis='both', which='major', labelsize=30, width=4)
plt.xticks(ticks=x, labels=x, rotation=75)
# plt.legend(fontsize=35, loc=(0.6, 0.25))

plt.savefig('time_train_data_versus_number_bins.png')
plt.show()
```
